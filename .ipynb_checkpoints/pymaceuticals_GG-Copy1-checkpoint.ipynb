```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load 
md_data = "data/mouse_drug_data.csv"
ct_data = "data/clinicaltrial_data.csv"
```

```python
# Read the Mouse-Drug Data 
md_df = pd.read_csv(md_data)
md_df.head()
```

```python
#Read the Clinical Trial Data
ct_df = pd.read_csv(ct_data)
ct_df.head()
```

```python
# Combine the data into a single dataset
merged_df = pd.merge(ct_df, md_df, on = "Mouse ID")
# Display the data table for preview
merged_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_df = merged_df.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]
# Convert to DataFrame
mean_df = pd.DataFrame(mean_df)
mean_df.head()
```

```python
# Preview DataFrame with reset index
mean_tvd = mean_df.reset_index(level=None, drop=False)
mean_tvd.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_df = merged_df.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]
# Convert to DataFrame
sem_df = pd.DataFrame(sem_df)
# Preview DataFrame
sem_tvd = sem_df.reset_index(level=None, drop=False)
sem_tvd.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformatted_df = mean_tvd.pivot_table(index='Timepoint', columns='Drug', values= 'Tumor Volume (mm3)')

# Preview that Reformatting worked
reformatted_df.head()
```

```python
#set x axis
x_axis = np.arange(0,50,5)
x_axis
```

```python
#set value for yerr
err_cap = sem_df["Tumor Volume (mm3)"]["Capomulin"]
err_inf = sem_df["Tumor Volume (mm3)"]["Infubinol"]
err_ket = sem_df["Tumor Volume (mm3)"]["Ketapril"]
err_plc = sem_df["Tumor Volume (mm3)"]["Placebo"]
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(x_axis, reformatted_df["Capomulin"], yerr=err_cap, fmt='o',ls='dashed', color='red', label="Capomulin")
plt.errorbar(x_axis, reformatted_df["Infubinol"], yerr=err_inf, fmt='^', ls='dashed', color='blue', label="Infubinol")
plt.errorbar(x_axis, reformatted_df["Ketapril"], yerr=err_ket, fmt='s', ls='dashed', color='green', label="Ketapril")
plt.errorbar(x_axis, reformatted_df["Placebo"], yerr=err_plc, fmt='d', ls='dashed', color='orange', label="Placebo")
plt.legend(loc='best')
plt.grid()
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.xlim(-2,47)
plt.ylim(33,74)
# Save the Figure
plt.savefig("output/Tumor Response to Treatment.png")
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_ms = merged_df.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]
# Convert to DataFrame
mean_ms = pd.DataFrame(mean_ms)
# Preview DataFrame
mean_mets = mean_ms.reset_index(level=None, drop=False)
mean_mets.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_ms = merged_df.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]
# Convert to DataFrame
sem_ms = pd.DataFrame(sem_ms)
# Preview DataFrame
sem_mets = sem_ms.reset_index(level=None, drop=False)
sem_mets.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_mets = mean_mets.pivot_table(index='Timepoint', columns='Drug', values= 'Metastatic Sites')

# Preview that Reformatting worked
reformat_mets.head()
```

```python
#set yerr for second graph (metastatic site data)
met_err_cap = sem_ms["Metastatic Sites"]["Capomulin"]
met_err_inf = sem_ms["Metastatic Sites"]["Infubinol"]
met_err_ket = sem_ms["Metastatic Sites"]["Ketapril"]
met_err_plc = sem_ms["Metastatic Sites"]["Placebo"]
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(x_axis, reformat_mets["Capomulin"], yerr=met_err_cap, fmt='o',
             ls='dashed',linewidth=0.5, color='red', label="Capomulin")
plt.errorbar(x_axis, reformat_mets["Infubinol"], yerr=met_err_inf, fmt='^', 
             ls='dashed',linewidth=0.5, color='blue', label="Infubinol")
plt.errorbar(x_axis, reformat_mets["Ketapril"], yerr=met_err_ket, fmt='s', 
             ls='dashed',linewidth=0.5, color='green', label="Ketapril")
plt.errorbar(x_axis, reformat_mets["Placebo"], yerr=met_err_plc, fmt='d', 
             ls='dashed',linewidth=0.5, color='orange', label="Placebo")
plt.legend(loc='best')
plt.grid()
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Metastatic Sites')
plt.xlim(-2,47)
plt.ylim(-0.2, 3.8)
# Save the Figure
plt.savefig("output/Metastatic Spread During Treatment.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mice_df = merged_df.groupby(["Drug", "Timepoint"]).count()["Mouse ID"].rename("Mouse Count")
# Convert to DataFrame
mice_df = pd.DataFrame(mice_df)
# Preview DataFrame
mice_count = mice_df.reset_index(level=None, drop=False)
mice_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_count = mice_count.pivot_table(index='Timepoint', columns='Drug', values= 'Mouse Count')

# Preview that Reformatting worked
reformat_count.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(x_axis, (reformat_count["Capomulin"]/25 * 100), marker='o',
             ls='dashed',linewidth=0.5, color='red', label="Capomulin")
plt.plot(x_axis, (reformat_count["Infubinol"]/25 * 100), marker='^', 
             ls='dashed',linewidth=0.5, color='blue', label="Infubinol")
plt.plot(x_axis, (reformat_count["Ketapril"]/25 * 100), marker='s', 
             ls='dashed',linewidth=0.5, color='green', label="Ketapril")
plt.plot(x_axis, (reformat_count["Placebo"]/25 * 100), marker='d', 
             ls='dashed',linewidth=0.5, color='orange', label="Placebo")
plt.legend(loc='best')
plt.grid()
plt.title('Survival During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate (%)')
plt.xlim(-2,48)
plt.ylim(32, 103)
# Save the Figure
plt.savefig("output/Survival During Treatment.png")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
# Display the data to confirm
pct_chnge = (reformatted_df.iloc[-1] - reformatted_df.iloc[0])/reformatted_df.iloc[0] * 100
pct_chnge
```

```python
drugs = ("Capomulin","Infubinol", "Ketapril", "Placebo")
tumor_vol_chnge = (-19.475303,46.123472, 57.028795, 51.297960)
x = np.arange(len(drugs))
ticks = [value for value in x]
colors =[]
for vol in tumor_vol_chnge:
    if vol >= 0:
        colors.append('red')
    else:
        colors.append('green')

plt.bar(x, tumor_vol_chnge, width =1, align='center', color=colors, linewidth=1, tick_label = drugs)
plt.grid()
plt.xlim(-0.75, len(x)-0.25)
plt.ylim(-30, 65)
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.xticks(ticks, drugs)
```
